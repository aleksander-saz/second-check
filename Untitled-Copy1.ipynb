```python
df2['age_cat'] = pd.qcut(df2['dob_years'],3,['молодежь','средний возраст','пожилые'])
df2['education'] = df2['education'].str.lower()
df2
```

```python
df2['total_income']=round(df2['total_income'])
df2.groupby(['age_cat','education','income_type'])['total_income'].transform('median').dropna().drop_duplicates().sort_values().reset_index(drop=1)
```
